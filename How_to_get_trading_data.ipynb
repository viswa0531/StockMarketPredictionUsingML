```python
import pandas as pd
```

```python
import quandl
```

```python
snp500url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

#scraping wikipedia to fetch S&P 500 stock list
data_table = pd.read_html(snp500url)
tickers = data_table[0].iloc[:, 0].tolist()
print("Total number of companies are: ",len(tickers))
```

```python
#fetching tickers(500) closing prices from google finance, using DataReader package for this purpose
price_list = []

for ticker in tickers:
    try:
        prices = quandl.get("WIKI/"+ticker, start_date="2017-01-01", end_date="2018-01-01", api_key="nUGzmRs2WcAHcSJf3NG2")['Close']
        prices = pd.DataFrame(prices)
        prices.columns = [ticker]
        price_list.append(prices)
        prices_df = pd.concat(price_list,axis=1)
    except: 
        pass
```

```python
prices_df.head()
```

```python
pd.DataFrame(prices_df).to_csv("/tmp/file.csv")
```
